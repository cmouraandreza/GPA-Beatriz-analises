**ANOSIM** = Analysis of Similarities.
Ela testa se as comunidades (composição de espécies) são mais parecidas dentro do mesmo grupo do que entre grupos diferentes.

Pensa assim:

* Você tem várias amostras (sites, pontos, meses…)

* Cada amostra tem abundâncias/presenças de espécies

* Você quer saber se “grupo A” tem comunidade diferente de “grupo B”

Ela é tipo um “ANOVA” só que usando distâncias ecológicas e permutação.

```r

ano = anosim(m_com, dados$Management, distance = "bray", permutations = 9999)
ano
```

**ANOSIM** precisa de réplicas por grupo para:

* funcionar corretamente

* ter poder estatístico

* fazer sentido ecológico

Com 1 amostra por grupo, o teste até pode rodar, mas o resultado não é confiável.
